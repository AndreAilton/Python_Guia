# **Estruturas de repetição**

## **While**
While True: <--- Repetição infinita
  if "Variavel" == 0:
    break <---- Condicional que quebra repetição

While "Variavel"== 0: <------ Repetição com condição
  "Escreva um codigo de condição" <---- Insira oque precisa ser repetido

Else: <------ Se a condição não for atendida realize isso
   "Escreva um codigo da condição" <---- Insira oque precisa ser repetido

```python
c = 0
while True:
  c += 1
  print(c)
  if c == 10:
    break
```

```python
c = 0
while c < 10:
  c += 1
  print(c)
```

```python
c = 0
while c < 10:
  c += 1
  print(c)
else:
  print(c+100)
```

## **for com range**
**for "Variavel" in range("Numero")**: <---"Insira a variavel" a Variavel vai ser igual a contagem da repetição
range(x)<----- Insira um numero que cria uma lista de 0 ao numero selecionado

```python
for c in range(11):#range com inico automatico
  print(c)
```

```python
for c in range(1,11):#range com inicio pre-estabelecido
  print(c)
```

## **for com Strings**
for "Variavel" in **"Texto"**: <----"Insira a variavel" a Variavel vai ser igual ao indice da lista
Texto <----- Insira uma string que vai ser tratado como uma lista e repetida ate o fim

```python
Lista = "Teste1 Teste2 Teste3 Teste4 Teste5"
for c in Lista:
  print(c)
```

## **for com Lista**
for "Variavel" in **"Lista"**: <---"Insira a variavel" a Variavel vai ser igual ao indice da lista
Lista <----- insira a lista que vai ser repetida

```python
Lista = ["Teste1", "Teste2", "Teste3", "Teste4", "Teste5"]
#-------------------------------
for c in Lista:
  print(c)
```

# **Estruturas De Condicionais**



## **If e Else**
if "condição": <---- Insira uma condição utilizando os operadores logicos como
if "variavel" > "condição":
  Se verdadeiro resultado:
else:
  Se Falso Resultado:

```python
x = int(input("Escreva um numero para comparar "))# Variavel para voce Comparar
if x > 9:
  print(x," e Maior do que", 9)
else:
  print(x," e menor do que ", 9)
```

## **If elif Else**
adicionando **Elif** na condição do if conseguimos colocar condições especificas diferente do else que se não se enquadradar no primeiro criterio o codigo **else** e executado:

Elif x == 9:

```python
x = int(input("Escreva um numero para comparar "))# Variavel para voce Comparar
if x > 9:
  print(x," e Maior do que", 9)
elif x == 9:
  print(x," e igual a ", 9)
else:
  print(x," e menor do que ", 9)
```

# **Formatação de String**

## **.strip() <--------- Remove os espaços em branco antes e depois do texto**

```python
Texto = "   Teste String    "
Texto = Texto.strip()
print(Texto)
```

## **.split()<------Transforma uma string em uma lista dividindo por espaços em branco**

```python
Texto = "Teste1 Teste2 Teste3 Teste4 Teste5"
Texto = Texto.split()
print(Texto)
```

## **.upper() .lower()<---------Coloca em maiusculo e minusculo**

```python
Texto = "Teste1 Teste2 Teste3 Teste4 Teste5".upper()
print(".Upper() ",Texto)
Texto = "Teste1 Teste2 Teste3 Teste4 Teste5".lower()
print(".lower() ",Texto)
```

##**.replace() <------- Troca um caractere de uma string por outro**

```python
Texto = "Teste1 Teste2 Teste3 Teste4 Teste5".upper()
Texto = Texto.replace("TESTE","Testado")
print(Texto)
```

```python
Texto = "Teste1 Teste2 Teste3 Teste4 Teste5".upper()
Texto = Texto.replace("TESTE","Testado",2)
print(Texto)
```

## **.format() <-------- Uma das principais funções ja que com ele você consegue inserir um texto formatado em uma string por meio dos {}**

```python
Texto = "{} {} {} {} {}".format('Teste1', 'Teste2', 'Teste3', 'Teste4', 'Teste5')
print(Texto)
```
